```python
from kafka import KafkaConsumer
```

```python
bootstrap_servers = ['localhost:9092']
```

```python
consumer = KafkaConsumer('my-topic', bootstrap_servers=bootstrap_servers, auto_offset_reset='earliest', group_id='my-group')
```

```python
for msg in consumer:
    print(f"Received message: {msg.value.decode('utf-8')}")
```

```python
consumer.close()
print("Consumer stopped")
```
